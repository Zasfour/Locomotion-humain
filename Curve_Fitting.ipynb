## Méthode avec ipopt

### Définissons nos fonctions.

```python
T1 = np.loadtxt("human_traj_1.dat")

X1 = T1[0+12]
X2 = T1[1+12]
```

```python
a = SX.sym('a',1)
b = SX.sym('b',1)
c = SX.sym('c',1)
d = SX.sym('d',1)
e = SX.sym('e',1)
f = SX.sym('f',1)
g = SX.sym('g',1)
h = SX.sym('h',1)
i = SX.sym('i',1)
j = SX.sym('j',1)
k = SX.sym('k',1)
l = SX.sym('l',1)


X = SX.sym('X',n)
Y = SX.sym('Y',n)


u = 0 
for m in range (n):
    u = u + (Y[m]-(a+ b*X[m]+ c*X[m]**2+ d*X[m]**3+ e*X[m]**4+ f*X[m]**5+ g*X[m]**6+ h*X[m]**7+ i*X[m]**8+ j*X[m]**9+ k*X[m]**10+ l*X[m]**11))**2
u = sqrt(u/n)

RSME = Function('RSME', [Y,X,a,b,c,d,e,f,g,h,i,j,k,l],[u])
```

### Polynôme de degré 3

```python
e = 0
f = 0
g = 0
h = 0
i = 0
j = 0
k = 0
l = 0

opti = casadi.Opti()

a = opti.variable(1)
b = opti.variable(1)
c = opti.variable(1)
d = opti.variable(1)


opti.minimize( RSME (X2,X1,a,b,c,d,e,f,g,h,i,j,k,l))

opti.subject_to(a + b*X1[0]+ c* X1[0]**2 + d* X1[0]**3  == X2[0])
opti.subject_to(b+ 2*c* X1[0] + 3*d* X1[0]**2  >= 0.01)
opti.subject_to(2*c + 3*2*d* X1[0]  >= 0.01 )

opti.subject_to(a + b*X1[-1]+ c* X1[-1]**2 + d* X1[-1]**3  == X2[-1])
opti.subject_to(b+ 2*c* X1[-1] + 3*d* X1[-1]**2 <= 0.01)
opti.subject_to(2*c + 3*2*d* X1[-1]  <= 0.01)


opti.solver('ipopt')     

sol = opti.solve()


Y3 = F3(X1,sol.value(a),sol.value(b),sol.value(c),sol.value(d))
```

### Polynôme de degré 4
